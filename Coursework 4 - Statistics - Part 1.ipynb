The paper emphasises that the p-value is often misunderstood. Some people believe that a p-value implies the probability that the null hypothesis is true (with the null hypothesis being the absence of any correlation between the number of storks and the number of live births in a particular country). 

For example they would say a p-value of 0.008 implies the storks giving birth to babies is the correct explanation in 1 out of 125 cases. This is false and it also doesnt imply that the alternative hypothesis that storks really do deliver babies is true 99.2% of the time.

The correct interpretation of the p-value in this paper should be that, given the null hypothesis is true (no correlation between storks and babies), there is a 1 out of 125 chance we would get a statistically significant value if we were to run the trial many times over. The correlation coefficient is not particularly high, but according to its p-value, there is only a 1 in 125 chance of obtaining at least as impressive a value assuming the null hypothesis of no correlation were true. This means a highly significant statistical correlation is present between the two variables, which is something which is clearly misinterpreted as it is not possible. The paper emphasises we must look beyond the face value of the p value and there are cases where there is no clear confounder or no clear non causality doesn't prove causality.

#### Question 2

The correlation coefficient describes a measure of how correlated variables are from -1 to 1 which describes the type of correlation (positive or negative) and how strong the correlation is. Two variables having correlation doesnt necessarily mean they are statistically significant which is a key difference between r and p values. p values are probability values and relate to how confident we are a parameter is statistically meaningful. A parameter can have a strong correlation but it may not mean anything if it isn't significant.

The p value is used to identify if our coefficients in our model are statistically significant. This relates to how good is my regression model because the p value will reveal the predictors which have an impact on the dependent variable. If all the predictors in a multivariable model are significant, then it is likely the model is a good fit and will produce good predictions with a high r2 and low error. If the p value shows that we reject our coefficients then the model is likely no good as the predictors have little to no effect. We use the p value this way by comparing it to our alpha value that we deem appropriate. 

If the p-value is smaller (the trial result is less likely) and the less likely the trial with null hypothesis true the more we can say our test result data didn't happen by random chance and was real. The smaller the p value the stronger the evidence you should reject the null hypothesis. The probability (p value) relates to running the trial many times, and how often we would get data as extreme / impressive just by chance.

#### Question 3

Causation describes a scenario where one variable directly impacts another variable, so they have a relationship where one is dependent on the other. Whereas correlation describes two variables having some sort of association. This could mean as A goes up so does B but this doesn't mean causality and it is important we intepret the p values correctly. There is a distinct difference where causality means one causes the other and correlation just means they have some sort of relationship but it is unclear.

These can easily be confused with each other, especially if people misinterpret what the p values are telling us. If it is believed that a p value of 0.008 means that the alternative hypothesis is true 99.2% of the time, then it may well be believed that the statistically significant value means one action causes the other action, e.g storks make babies. To prevent making these mistakes we must look deeper into what is causing low p-value. 

Although correlation doesn't imply causation, its absence would certainly constitute evidence against a simple
relationship. This shows that we can prove there is no causation if there is no correlation, so the two do have a relationship in this way. We just can't prove there is causation with correlation.

The problem is, although a high correlation should not be taken as causation, there are some examples where it is not obviously only a mere association such as the apparent relationship between educational level and unemployment. There are also examples where high correlation is easily explained by a confounding variable such as the correlation between the reading skills of children and their shoe size study. However the storks and babies case study is a particularly rare case where there is no clear confounding variable to dismiss the relationship so it can be hypothesised that there is actually more than just an 'association', which in reality is ridiculous. This emphasises the fact we can't always rely on the fact that there is a significant effect from the p-value.
